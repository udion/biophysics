```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import os, sys
```

```python
with open('../data/frapdata1.dat') as f:
    lines = f.readlines()
lines1  = [l.strip('\n').split() for l in lines]
time = []
vals = []
for p in lines1:
    if p != []:
        time.append(float(p[0]))
        vals.append(float(p[1]))
# print(lines)
plt.figure(figsize=(20,10))
plt.scatter(time, vals)
```

```python
def Nft(t, D):
    a = 0.5 * (10**-6)
    L = 2.5 * (10**-6)
    s = 0
    for i in range(1,200):
        s = s+ (2/(i*np.pi)**2)*(np.sin((i*np.pi*a)/(L))**2)*(np.exp((-1*D*t*(i*np.pi)**2)/(L**2)))
    s = (1/((a/L)*(1-(a/L))))*s
    return 1-s
```

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(Nft, time, vals)
```
